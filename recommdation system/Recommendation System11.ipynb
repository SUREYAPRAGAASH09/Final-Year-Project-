```python
def generateGenreMatrixTrain(data):
    genreMatrix=[]
    for i in data:
        tempGM=[]
        genList=i.split('|')
        for j in genreList:
            if j in genList:
                tempGM.append(1)
            else:
                tempGM.append(0)
        
        tempGM[-1]=genList[-1]
        genreMatrix.append(tempGM)
    return genreMatrix

def generateGenreMatrixTest(data):
    genreMatrix=[]
    for mid,i in data.values:
        tempGM=[mid]
        genList=i.split('|')
        for j in genreList:
            if j in genList:
                tempGM.append(1)
            else:
                tempGM.append(0)
        
        genreMatrix.append(tempGM)
    return genreMatrix
    
```

```python
genreList=['Action','Adventure','Animation','Children''s','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western','Rating']

# Find movies rated by user - UID 1
UID=1
userMovieList=ratings[ratings.UserID==UID].drop('Timestamp',axis=1)
userMovieGenre=pd.merge(userMovieList,movies,on='MovieID',how='left')

userMovieGenre['RatingStr']=userMovieGenre['Rating'].map(lambda a: str(a))
userMovieGenre['GenreRating']=userMovieGenre['Genres']+"|"+userMovieGenre['RatingStr']

# Find trainMatrix
trainMatrix=pd.DataFrame(generateGenreMatrixTrain(userMovieGenre['GenreRating']),columns=genreList)
trainMatrix
```

```python
#Find Similar Users
otherUserMovieList=ratings[ratings.UserID!=UID].drop('Timestamp',axis=1)
userVsOtherUser=pd.merge(userMovieList,otherUserMovieList,on='MovieID')

otherUserRatingPT=pd.pivot_table(userVsOtherUser,index=['UserID_y'],columns=['MovieID'],values='Rating_y')

otherUsersID=otherUserRatingPT.index
correlation=[]
for other in otherUsersID:
    commonRatedMovies=userVsOtherUser[userVsOtherUser['UserID_x']==UID][userVsOtherUser['UserID_y']==other]
    commonRatings=commonRatedMovies[['Rating_x','Rating_y']]

    correlation.append([commonRatings['Rating_x'].corr(commonRatings['Rating_y'],method='pearson')])

corrMat=pd.DataFrame(correlation,index=otherUsersID,columns=['Correlation'])

selectedOtherUsers=corrMat[corrMat['Correlation']>.5]

selectedOtherUsers
```

```python
corrMat.head()
```

```python
#Adding Similar User Details to TrainMatrix

for i in selectedOtherUsers.index:
    userMovieList=ratings[ratings.UserID==i].drop('Timestamp',axis=1)
    userMovieGenre=pd.merge(userMovieList,movies,on='MovieID',how='left')

    userMovieGenre['RatingStr']=userMovieGenre['Rating'].map(lambda a: str(a))
    userMovieGenre['GenreRating']=userMovieGenre['Genres']+"|"+userMovieGenre['RatingStr']

    # Find trainMatrix
    trainMatrix=trainMatrix.append(pd.DataFrame(generateGenreMatrixTrain(userMovieGenre['GenreRating']),columns=genreList))
```

```python
genreListTest=['MovieID','Action','Adventure','Animation','Children''s','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western','Rating']
test=pd.DataFrame(generateGenreMatrixTest(movies[['MovieID','Genres']]),columns=genreListTest)
test
```

```python
from sklearn import ensemble
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split


X_train = trainMatrix.drop('Rating',axis=1).values
y_train=trainMatrix['Rating'].values

X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2,random_state=9)

X_test=test.drop(['MovieID','Rating'],axis=1).values

# Fit classifier with out-of-bag estimates
params = {'n_estimators': 1200, 'max_depth': 3, 'subsample': 0.5,
          'learning_rate': 0.01, 'min_samples_leaf': 1, 'random_state': 3}

clf = ensemble.GradientBoostingRegressor(**params)

clf.fit(X_train, y_train)

acc = clf.predict(X_test)

predictions=pd.concat([pd.DataFrame(test['MovieID']),pd.DataFrame(acc,columns=['PredRating'])],axis=1)
```
